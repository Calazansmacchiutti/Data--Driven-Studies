```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'S' is already a diagonal matrix with singular values obtained from np.linalg.svd

# Plot the singular values on a semilogarithmic scale
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()

# Plot the cumulative sum of the singular values normalized by the total sum of the singular values
plt.figure(2)
plt.plot(np.cumsum(np.diag(S)) / np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()
```
